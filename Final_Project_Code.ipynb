# Classification metrics
Evaluate the expected loss or gain for an investor deciding, whether to invest into a company based on the classification into bankrupt or non bankrupt

```python
# define cost functions for expected loss of investment strategy
# the expected gain is the negative of the expected loss
# NOTE: the formulas are derived in the report 

# probability of actual bankruptcy P(B)
# under the assumption that the dataset is representative for the actual distribution of bankruptcies
p_bankrupt = class_counts[1]/sum(class_counts)
print("P(B) =", p_bankrupt)

from sklearn.metrics import confusion_matrix

# use dictionary output of classification_report to determine expected loss, when investing based on the classifier
def cost_classifier(y_test, y_pred):
    # positive hypothesis: bankruptcy
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    return (fn - 0.5 * tn)/(tn+fp+fn+tp)

# determine expected loss, when always investing
def cost_blind_invest():
    return -0.5 + 1.5 * p_bankrupt

# determine expected loss, when never investing
def cost_blind_not_invest():
    return 0

# determine how much additional gain in return on investment is made by using the classifier
# instead of any blind baseline investment strategy
def additional_gain_through_classifier(y_test, y_pred):
    gain_classifier = -cost_classifier(y_test, y_pred)
    gain_blind_invest = -cost_blind_invest()
    gain_blind_not_invest = -cost_blind_not_invest()
    return min(gain_classifier - gain_blind_invest, gain_classifier - gain_blind_not_invest)

# determine what fraction of possible additional gain in return on investment is made by using the classifier
# instead of any blind baseline investment strategy
def fraction_of_possible_gain_through_classifier(y_test, y_pred):
    return additional_gain_through_classifier(y_test, y_pred) / additional_gain_through_classifier(y_test, y_test)
```

# Dimensionality Reduction and Hyper Parameter Search

We select different ML Algorithms, different dimensionality reduction techniques, different hyperparameters to produce the best models.

Note: this implementation of the hyper parameter optimisation is very slow especially for some of the hyperparameter instances involving SVMs

```python
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.metrics import make_scorer, confusion_matrix
from sklearn.decomposition import PCA
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.feature_selection import SelectKBest, f_classif, chi2, mutual_info_classif
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import numpy as np
```

## Functions to help perform the dimensionality reduction

```python
# select set of features with an absolute correlation above threshold
def corr_limit(dataset, threshold):
    col_corr = set() # Set of all the names of deleted columns
    corr_matrix = dataset.corr().abs()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j] >= threshold) and (corr_matrix.columns[j] not in col_corr):
                colname = corr_matrix.columns[i] # getting the name of column
                col_corr.add(colname)
    return col_corr
    

# Purpose: Select the most important features from the dataset to improve model performance
# turn the whole dataset into (X_train, X_test, y_train, y_test) using feature selection that is trained on training data
def select_features(data, correlation_limit, feature_selection_algorithm, feature_selection_count):
    #Step 1: Separate the features and the target variable
    features = data.drop('Bankrupt?', axis=1)  
    target = data['Bankrupt?']

    #Step 2: Perform train-test split
    X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42, stratify=y)

    #Step 3: Perform feature selection

    # remove n-1 of n features with very high correlation - deactivate by setting correlation_limit to 1.1
    col_corr = corr_limit(X_train, correlation_limit)
    X_train = X_train.drop(columns=col_corr)
    X_test = X_test.drop(columns=col_corr)

    # use different feature selection functions from sklearn.feature_selection
    if 'sklearn.feature_selection' in feature_selection_algorithm.__module__:
        kBest = SelectKBest(score_func=feature_selection_algorithm, k=feature_selection_count).fit(X_train, y_train)
        X_train = kBest.transform(X_train)
        X_test = kBest.transform(X_test)

    # use RandomForestClassifier to select the features
    elif 'RandomForestClassifier' in feature_selection_algorithm.__name__:
        # fit the classifier
        clf = RandomForestClassifier()
        clf.fit(X_train, y_train)

        #Get feature importance scores from the trained Random Forest model
        feature_importance = clf.feature_importances_

        #Sort the features based on their importance scores in descending order
        sorted_indices = np.argsort(feature_importance)[::-1]

        #Select the top k features based on their importance scores
        selected_features = X_train.columns[sorted_indices[:feature_selection_count]]
        X_train = X_train[selected_features]
        X_test = X_test[selected_features]
        
    # support for PCA as a feature extraction method
    elif 'PCA' in feature_selection_algorithm.__name__:
        pca = PCA(n_components=feature_selection_count)
        pca.fit(X_train, y_train)
        X_train = pca.transform(X_train)
        X_test = pca.transform(X_test)
    
    return (X_train, X_test, y_train, y_test)
```

## Hyperparameter search
1. Write sets of possible hyperparameters
2. For each type of classifier:
    1. Iterate over all hyperparameters
    2. Train model with specific hyperparameters
    3. Evaluate the score of the model
    4. If the model performed best for the type of classifier: save the hyperparameters and scores

```python
### (Fully) tuned Models: MLPClassifier, RandomForestClassifier, GradientBoostingClassifier
### Needing tuning: BernoulliNB, SVC (runs very slowly), KNeighborsClassifier, DecisionTreeClassifier

# types of classifiers
arr_classifier = {RandomForestClassifier, GradientBoostingClassifier, MLPClassifier} #, {GradientBoostingClassifier, MLPClassifier, BernoulliNB, RandomForestClassifier, SVC, KNeighborsClassifier, DecisionTreeClassifier}

# since every type of classifier is different, it has a different optimal set of hyperparameters, which need to be tuned:
# tuning: start off with a wide range of parameters, close in on the best performing parameters
arr_correlation_limit = { # {0.8, 0.9, 0.95, 0.97, 0.99, 1}
    MLPClassifier : {1}, 
    BernoulliNB : {0.8, 0.9, 0.95, 0.97, 0.99, 1}, 
    RandomForestClassifier : {1},
    GradientBoostingClassifier : {1},
    SVC : {0.8, 0.9, 0.95, 0.97, 0.99, 1},
    KNeighborsClassifier : {0.8, 0.9, 0.95, 0.97, 0.99, 1},
    DecisionTreeClassifier : {0.8, 0.9, 0.95, 0.97, 0.99, 1}
}
arr_feature_selection_algorithm = { # {PCA, RandomForestClassifier, f_classif, chi2, mutual_info_classif}
    MLPClassifier : {mutual_info_classif}, 
    BernoulliNB : {PCA, RandomForestClassifier, chi2, mutual_info_classif}, 
    RandomForestClassifier : {RandomForestClassifier},
    GradientBoostingClassifier : {RandomForestClassifier},
    SVC : {PCA, RandomForestClassifier, chi2, mutual_info_classif},
    KNeighborsClassifier : {PCA, RandomForestClassifier, chi2, mutual_info_classif},
    DecisionTreeClassifier : {PCA, RandomForestClassifier, chi2, mutual_info_classif}
}
arr_feature_selection_count =  { # {10, 20, 35, 50}  # (35 seemed best for f_classif and BernoulliNB)
    MLPClassifier : {20}, 
    BernoulliNB : {10, 20, 35, 50}, 
    RandomForestClassifier : {10},
    GradientBoostingClassifier : {17},
    SVC : {10},
    KNeighborsClassifier : {10, 20, 35, 50},
    DecisionTreeClassifier : {10, 20, 35, 50}
}

# Define the parameter grids for each classifier
param_grid = {
    MLPClassifier: {
        'solver': ['lbfgs'],
        'hidden_layer_sizes': [(69)],
        'max_iter': [800]
    },
    KNeighborsClassifier: {
        'n_neighbors': [1,3,5,7,10],
        'weights': ['uniform', 'distance'],
        'p': [1,2],
        'algorithm': ['ball_tree', 'kd_tree', 'brute']
    },
    DecisionTreeClassifier: {
        'criterion': ['gini', 'entropy'],
        'max_depth': [None, 5, 10],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
        'class_weight': [None, 'balanced']
    },
    RandomForestClassifier: {
        'n_estimators': [130], # [20, 50, 100, 150],
        'max_depth': [None], # [None, 5, 10]
        'min_samples_split': [2],
        'min_samples_leaf': [2],
        'class_weight': ['balanced'] # [None, 'balanced']
    },
    GradientBoostingClassifier: {
        'n_estimators': [90],
        'learning_rate': [0.1],
        'max_depth': [None], # [None, 5, 10],
        'min_samples_split': [2], # [2, 5, 10],
        'min_samples_leaf': [4] # [1, 2, 4]
    },
    BernoulliNB: {
        'alpha': [0.1, 0.5, 1.0],
        'binarize': [0.0, 0.5, 1.0]
    },
    SVC: {
        'C': [0.1], #[0.1, 1, 10, 100],
        'kernel': ['linear'], # ['linear', 'rbf', 'poly', 'sigmoid'],
        'gamma': ['scale'] , # ['scale', 'auto'],
        'class_weight': ['balanced'] # [None, 'balanced']
    }
}

# iterate over all classifiers
for classifier in arr_classifier: #{RandomForestClassifier, BernoulliNB, SVC, KNeighborsClassifier, DecisionTreeClassifier}
    print("Now trying classifier:", classifier.__name__)
    best_score = -100
    
    # find best hyperparameters for preprocessing
    for correlation_limit in arr_correlation_limit[classifier]: # 0.97 {0.8, 0.9, 0.95, 0.97, 0.99, 1}
        for feature_selection_algorithm in arr_feature_selection_algorithm[classifier]: # {RandomForestClassifier, f_classif, chi2, mutual_info_classif}
            for feature_selection_count in arr_feature_selection_count[classifier]: # {10, 20, 35, 50}  (35 seemed best for f_classif)
                # select features according to hyperparameters
                X_train_reduced, X_test_reduced, y_train, y_test = select_features(new_data, correlation_limit, feature_selection_algorithm, feature_selection_count)

                # Perform grid search with cross-validation for each classifier across all possible parameters for the classifier from param_grid
                # the scoring function used is fraction_of_possible_gain_through_classifier as it represents how usefull the model is
                classifier_instance = classifier()
                classifier_grid_search = GridSearchCV(classifier_instance, param_grid[classifier], cv=5, scoring=make_scorer(score_func=fraction_of_possible_gain_through_classifier))
                classifier_grid_search.fit(X_train_reduced, y_train) # training data
                
                # Get the best hyperparameters from param_grid for current classifier and current preprocessing settings
                best_classifier_params = classifier_grid_search.best_params_
                score = classifier_grid_search.best_score_
                print("Best", classifier.__name__, "Hyperparameters:", best_classifier_params, "Score:", score)

                # incase the specific choice of preprocessor hyperparameters has better score, overwrite entry in file
                if score > best_score:
                    best_score = score # score is fraction_of_possible_gain_through_classifier
                    best_classifier_pred = classifier_grid_search.best_estimator_.predict(X_test_reduced)

                    # calculate precision scores
                    tn, fp, fn, tp = confusion_matrix(y_test, best_classifier_pred).ravel()
                    positive_precision = tp / (tp + fp)
                    negative_precision = tn / (tn + fn)

                    # load old information on hyperparameter and performance of classifier
                    try:
                        with open('hyperparam_dict.pickle', 'rb') as f:
                            hyperparam_dict = pickle.load(f)
                    except:
                        hyperparam_dict = {}

                    # save new information on hyperparameters and performance of classifier
                    hyperparam_dict[classifier.__name__] = {
                        'correlation_limit': correlation_limit,
                        'feature_selection_algorithm': feature_selection_algorithm,
                        'feature_selection_count': feature_selection_count,
                        'score': score,
                        'positive_precision': positive_precision,
                        'negative_precision': negative_precision,
                        'params': classifier_grid_search.best_params_
                    }

                    # save hyperparameters to file
                    with open('hyperparam_dict.pickle', 'wb') as f:
                        pickle.dump(hyperparam_dict, f)

                    print(f"fraction of possible additional gain using classification instead of blindly investing: {score}")
```

# Our best model

We combine the best models and use a kind of Voting-algorithm to determine the classification, however we have some minor changes to the way the votes are weighted and evaluated:
1. Every model votes for the probability of a given company being bankrupt, based on the probability of the company being bankrupt given their classification: P(B|f(x)), where B is bankrupt and f(x) is the classification
2. The mean bankruptcy probability from each models is calculated
3. The classification is performed by checking if the mean bankruptcy probability is beyond a tuneable threshold

The tuneable threshold serves the purpose of trading type 1 for type 2 errors in order to optimize the balance

## Train

```python
# set of classifiers being used
arr_classifier = {RandomForestClassifier, GradientBoostingClassifier, MLPClassifier}

# instances of models
models = {}
test_data = {}

for classifier in arr_classifier:
    # extract hyperparameters from dictionary
    correlation_limit = hyperparam_dict[classifier.__name__]['correlation_limit']
    feature_selection_algorithm = hyperparam_dict[classifier.__name__]['feature_selection_algorithm']
    feature_selection_count = hyperparam_dict[classifier.__name__]['feature_selection_count']
    params = hyperparam_dict[classifier.__name__]['params']
    positive_precision = hyperparam_dict[classifier.__name__]['positive_precision']
    negative_precision = hyperparam_dict[classifier.__name__]['negative_precision']
    
    # preprocess data according to best hyperparameters
    X_train_reduced, X_test_reduced, y_train, y_test = select_features(new_data, correlation_limit, feature_selection_algorithm, feature_selection_count)

    # train
    classifier_instance = classifier(**params)
    classifier_instance.fit(X_train_reduced, y_train)
    
    # save model, input for later testing
    models[classifier.__name__] = classifier_instance
    test_data[classifier.__name__] = X_test_reduced
```

## Test and Evaluate

```python
# add up the probabilities for bankruptcy
sum_chance_of_bankruptcy = np.zeros(len(y_test))

# minimum chance of bankruptcy needed to be classified as such, tune in order to maximize the score
min_chance_of_bankruptcy = 0.1

for classifier in arr_classifier:
    # test
    y_pred = models[classifier.__name__].predict(test_data[classifier.__name__])
    
    # add the chance of bankruptcy
    ## chance of bankruptcy given bankruptcy classification: P(b|c) = positive_precision
    sum_chance_of_bankruptcy[y_pred==1] += positive_precision 
    ## chance of bankruptcy given non-bankruptcy classification: P(b|not-c) = 1 - P(not-b|not-c) = 1 - negative_precision
    sum_chance_of_bankruptcy[y_pred==0] += 1 - negative_precision

# average chance of bankruptcy
chance_of_bankruptcy = sum_chance_of_bankruptcy / len(arr_classifier)

y_pred_combined = np.zeros_like(y_pred)
# if the average chance of bankruptcy is greater than the minimum threshold then the datapoint is classified as bankrupt
y_pred_combined[chance_of_bankruptcy >= min_chance_of_bankruptcy] = 1

# statistics
score = fraction_of_possible_gain_through_classifier(y_test, y_pred_combined)
tn, fp, fn, tp = confusion_matrix(y_test, y_pred_combined).ravel()
positive_precision = tp / (tp + fp)
negative_precision = tn / (tn + fn)
print("score: ", score)
print("positive_precision: ", positive_precision)
print("negative_precision: ", negative_precision)
```
