# Compare ML Algorithms

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import classification_report


# Split the dataset into features and target
X = new_data.iloc[:, 1:]  # Select all rows and columns starting from index 1 to the end
y = new_data.iloc[:, 0]  # Select all rows from the first column
```

```python
# Count the occurrences of each class
class_counts = y.value_counts()

# Display the counts
print(class_counts)
print(max(class_counts)/min(class_counts))
```

```python
# define cost functions for expected loss of investment strategy
# the expected gain is the negative of the expected loss
# NOTE: the formulas are derived in the report 

# probability of actual bankruptcy P(B)
# under the ussumption that the dataset is representative for the actual distribution of bankruptcies
p_bankrupt = class_counts[1]/sum(class_counts)
print("P(B) =", p_bankrupt)

# use dictionary output of classification_report to determine expected loss, when investing based on the classifier
def cost_classifier(y_test, y_pred):
    report = classification_report(y_test, y_pred, output_dict=True)
    # P(B, \overline{C}) = P(\overline{C} | B) * P(B) = (1 - P(C|B)) * P(B) = (1 - recall(B)) * P(B)
    try:
        p_bankrupt_not_classified = (1 - report['1']['recall']) * p_bankrupt
    except:
        p_bankrupt_not_classified = (1 - report[1]['recall']) * p_bankrupt
    
    # P(\overline{B}, \overline{C}) = P(\overline{C} | \overline{B}) * P(\overline{B}) =
    # = recall(\overline{B}) * (1-P(b))
    try:
        p_not_bankrupt_not_classified = report['0']['recall'] * (1-p_bankrupt)
    except:
        p_not_bankrupt_not_classified = report[0]['recall'] * (1-p_bankrupt)
    
    return p_bankrupt_not_classified - 0.5 * p_not_bankrupt_not_classified
    
# determine expected loss, when always investing
def cost_blind_invest():
    return -0.5 + 1.5 * p_bankrupt

# determine expected loss, when never investing
def cost_blind_not_invest():
    return 0

# determine how much additional gain in return on investment is made by using the classifier
# instead of any blind baseline investment strategy
def additional_gain_through_classifier(y_test, y_pred):
    gain_classifier = -cost_classifier(y_test, y_pred)
    gain_blind_invest = -cost_blind_invest()
    gain_blind_not_invest = -cost_blind_not_invest()
    return min(gain_classifier - gain_blind_invest, gain_classifier - gain_blind_not_invest)

# determine what fraction of possible additional gain in return on investment is made by using the classifier
# instead of any blind baseline investment strategy
def fraction_of_possible_gain_through_classifier(y_test, y_pred):
    return additional_gain_through_classifier(y_test, y_pred) / additional_gain_through_classifier(y_test, y_test)
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=43, stratify=y)
```

Once we have our data split into training and testing sets, we are going to evaluate the different ML Algorithms.
KNN, Decision Trees, Bayes Classifiers (we will choose Bernoulli Naive Bayes, which is suitable for our case because our target feature is binary (represented by 0 or 1)) and SVM.


```python
# Initialize the classifiers
knn = KNeighborsClassifier()
dt = DecisionTreeClassifier()
bnb = BernoulliNB()
bnb_reduced = BernoulliNB()
lr = LogisticRegression(max_iter=10000)
svm = SVC()

# feature selection can improve the performance of the naive Bayes by improving additional gain by factor of 2
from sklearn.feature_selection import SelectKBest
kBest = SelectKBest(k=35).fit(X_train, y_train)
X_train_reduced = kBest.transform(X_train)
X_test_reduced = kBest.transform(X_test)

# Train the classifiers
knn.fit(X_train, y_train)
dt.fit(X_train, y_train)
bnb.fit(X_train, y_train)
bnb_reduced.fit(X_train_reduced, y_train)
lr.fit(X_train, y_train)
svm.fit(X_train, y_train)

# Make predictions
knn_pred = knn.predict(X_test)
dt_pred = dt.predict(X_test)
bnb_pred = bnb.predict(X_test)
bnb_pred_reduced = bnb_reduced.predict(X_test_reduced)
lr_pred = lr.predict(X_test)
svm_pred = svm.predict(X_test)

# Compute the classification report
knn_report = classification_report(y_test, knn_pred)
dt_report = classification_report(y_test, dt_pred)
bnb_report = classification_report(y_test, bnb_pred)
bnb_report_reduced = classification_report(y_test, bnb_pred_reduced)
lr_report = classification_report(y_test, lr_pred)
svm_report = classification_report(y_test, svm_pred)

print("K-Nearest Neighbors Classification Report:")
print(knn_report)
print(f"gain through classification: {-cost_classifier(y_test, knn_pred)}")
print(f"gain blindly investing: {-cost_blind_invest()}")
print(f"fraction of possible additional gain using classification instead of blindly investing: {fraction_of_possible_gain_through_classifier(y_test, knn_pred)}")

print("\nDecision Tree Classification Report:")
print(dt_report)
print(f"fraction of possible additional gain using classification instead of blindly investing: {fraction_of_possible_gain_through_classifier(y_test, dt_pred)}")

print("\nBernoulli Naive Bayes Classification Report:")
print(bnb_report)
print(f"fraction of possible additional gain using classification instead of blindly investing: {fraction_of_possible_gain_through_classifier(y_test, bnb_pred)}")
print(bnb_report_reduced)
print(f"fraction of possible additional gain using classification instead of blindly investing: {fraction_of_possible_gain_through_classifier(y_test, bnb_pred_reduced)}")

print("\nLogistic Regression Classification Report:")
print(lr_report)
print(f"fraction of possible additional gain using classification instead of blindly investing: {fraction_of_possible_gain_through_classifier(y_test, lr_pred)}")

print("\nSupport Vector Machines Classification Report:")
print(svm_report)
print(f"fraction of possible additional gain using classification instead of blindly investing: {fraction_of_possible_gain_through_classifier(y_test, svm_pred)}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

#Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(selected_data, new_data['Bankrupt?'], test_size=0.3, random_state=42, stratify=y)
#it is trained on selected data which is fore selected features


# Define the parameter grids for each classifier
knn_param_grid = {
    'n_neighbors': [1,3,5,7,10],
    'weights': ['uniform', 'distance'],
    'p': [1,2],
    'algorithm': ['ball_tree', 'kd_tree', 'brute']
}

dt_param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'class_weight': [None, 'balanced']
}

bnb_param_grid = {
    'alpha': [0.1, 0.5, 1.0],
    'binarize': [0.0, 0.5, 1.0]
}

svm_param_grid = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'rbf', 'poly', 'sigmoid'],
    'gamma': ['scale', 'auto'],
    'class_weight': [None, 'balanced']
}

# Perform grid search with cross-validation for each classifier
knn = KNeighborsClassifier()
knn_grid_search = GridSearchCV(knn, knn_param_grid, cv=5)
knn_grid_search.fit(X_train, y_train)

dt = DecisionTreeClassifier()
dt_grid_search = GridSearchCV(dt, dt_param_grid, cv=5)
dt_grid_search.fit(X_train, y_train)

bnb = BernoulliNB()
bnb_grid_search = GridSearchCV(bnb, bnb_param_grid, cv=5)
bnb_grid_search.fit(X_train, y_train)

svm = SVC()
svm_grid_search = GridSearchCV(svm, svm_param_grid, cv=5)
svm_grid_search.fit(X_train, y_train)

# Get the best hyperparameters for each classifier
best_knn_params = knn_grid_search.best_params_
best_dt_params = dt_grid_search.best_params_
best_bnb_params = bnb_grid_search.best_params_
best_svm_params = svm_grid_search.best_params_

print("Best KNN Hyperparameters:", best_knn_params)
print("Best Decision Trees Hyperparameters:", best_dt_params)
print("Best Bernoulli Naive Bayes Hyperparameters:", best_bnb_params)
print("Best SVM Hyperparameters:", best_svm_params)

# Use the best hyperparameters to train the final classifiers
best_knn = KNeighborsClassifier(**best_knn_params)
best_dt = DecisionTreeClassifier(**best_dt_params)
best_bnb = BernoulliNB(**best_bnb_params)
best_svm = SVC(**best_svm_params)

best_knn.fit(X_train, y_train)
best_dt.fit(X_train, y_train)
best_bnb.fit(X_train, y_train)
best_svm.fit(X_train, y_train)

# Evaluate the performance of the final classifiers
best_knn_pred = best_knn.predict(X_test)
best_dt_pred = best_dt.predict(X_test)
best_bnb_pred = best_bnb.predict(X_test)
best_svm_pred = best_svm.predict(X_test)

# Compute the classification report
best_knn_report = classification_report(y_test, best_knn_pred)
best_dt_report = classification_report(y_test, best_dt_pred)
best_bnb_report = classification_report(y_test, best_bnb_pred)
best_svm_report = classification_report(y_test, best_svm_pred)

print("K-Nearest Neighbors Classification Report:")
print(best_knn_report)
print(f"fraction of possible additional gain using classification instead of blindly investing: {fraction_of_possible_gain_through_classifier(y_test, best_knn_pred)}")

print("\nDecision Tree Classification Report:")
print(best_dt_report)
print(f"fraction of possible additional gain using classification instead of blindly investing: {fraction_of_possible_gain_through_classifier(y_test, best_dt_pred)}")

print("\nBernoulli Naive Bayes Classification Report:")
print(best_bnb_report)
print(f"fraction of possible additional gain using classification instead of blindly investing: {fraction_of_possible_gain_through_classifier(y_test, best_bnb_pred)}")

print("\nSupport Vector Machines Classification Report:")
print(best_svm_report)
print(f"fraction of possible additional gain using classification instead of blindly investing: {fraction_of_possible_gain_through_classifier(y_test, best_svm_pred)}")
```
